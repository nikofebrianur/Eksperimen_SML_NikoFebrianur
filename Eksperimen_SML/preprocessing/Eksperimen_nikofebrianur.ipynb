##Menghapus Kolom Tidak Relevan

```python
df.drop(columns=['brand', 'sku', 'url', 'Product ID', 'scraped_at', 'currency'], inplace=True, errors='ignore')
df.info()
```

##Menangani Missing Values


```python
df.fillna(method='ffill', inplace=True)
df.isnull().sum()
```

##Encoding Label Kategorikal dan Scaling Fitur Numerik

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# --- 1. Pisahkan kolom kategorikal dan numerikal dengan benar ---
# Kategorikal: semua object
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Numerikal: hanya kolom yang memang float/int + variasi > 1
numerical_cols = ['price', 'Sales Volume']  # hardcoded berdasarkan df.info()

# --- 2. Label Encoding untuk kategorikal ---
le = LabelEncoder()
for col in categorical_cols:
    if df[col].nunique() > 1:
        df[col] = le.fit_transform(df[col])
    else:
        df.drop(columns=col, inplace=True)  # drop yang 1 value

# --- 3. Scaling untuk numerikal ---
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df.to_csv('/content/zara_preprocessed.csv', index=False)
```
